Agregamos un punto final "." a cada valor de *status* por que queremos agrupar estos resultados por constructores entonces esto nos facilitará el proceso. Ahora haremos uso de Siuba para obtener el DataFrame deseado.

```python
from siuba import * 
from siuba.dply.vector import *
```

Agruparemos el dataframe por año, ronda, carrera y constructor posteriormente sumaremos los puntos de los constructores en cada una de ellas y el status de sus 2 conductores. 
Despues los pasos realizados son cambiar el formato de presentacion de status separandolos con un "split" pero como los mostrará como una lista y se ve mejor como cadena vectorizamos la función "join" para cada uno de los valores de status.

```python
df=constructorResults2 >> group_by(_.year,_.round,_.raceId,_.constructorId)  >> summarize(points=_.points.sum(),status=_.status.sum())>>arrange (_.year,_.round)
df["status"]=df["status"].str.split(".",n=1)
df["status"]=[", ".join(x) for x in df["status"]]
df
```
